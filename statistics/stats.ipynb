```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_bar(data, Title):
    labels = list(data.keys())
    values = list(data.values())

    # Create the plot using seaborn's barplot function
    sns.set_style("whitegrid")  # Use a clean and modern theme
    plt.figure(figsize=(6, 4))  # Set the figure size

    sns.barplot(x=labels, y=values)

    # Customize the appearance of the plot
    plt.title(Title, fontsize=18)
    plt.xlabel("Proximity to Original Content", fontsize=14)
    plt.ylabel("Count", fontsize=14)
    plt.xticks(rotation=45)  # Rotate the x-axis tick labels for better readability
    plt.tight_layout()  # Add some padding to the plot

    # Show the plot
    plt.show()


```

```python
plot_bar(Pizza, "Pizza")
```

```python
plot_bar(SVP, "SVP Themed")
```

```python
plot_bar(German_Right_Wing, "German Right Wing")
```

```python
plot_bar(Gun_type, "Gun Type")
```

```python
plot_bar(Left, "Left Wing")
```

```python
plot_bar(Minecraft, "Let's Plays and Minecraft")
```

```python
plot_bar(Gronkh_TTT, "Gronkh")
```
